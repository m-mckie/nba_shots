```python
# want to make a height difference column
# for player
df['feet'], df['inches'] = df['HEIGHT'].str.split('-', 1).str

df['feet'] = pd.to_numeric(df['feet'])
df['inches'] = pd.to_numeric(df['inches'])

# turning into inches
df['INCHES'] = (df['inches']) + ((df['feet']) * 12)

# for defender
df['feet1'], df['inches1'] = df['CLOSE_DEF_HEIGHT'].str.split('-', 1).str

df['feet1'] = pd.to_numeric(df['feet1'])
df['inches1'] = pd.to_numeric(df['inches1'])

# turning into inches
df['DEF_INCHES'] = (df['inches1']) + ((df['feet1']) * 12)

# getting the difference
df['HEIGHT_DIFFERENCE'] = df['INCHES'] - df['DEF_INCHES']
```

```python
# now weight difference
df['WEIGHT_DIFFERENCE'] = df['WEIGHT'] - df['CLOSE_DEF_WEIGHT']
```
